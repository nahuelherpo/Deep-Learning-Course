```python
# Load dataset
df_seeds = pd.read_csv('data/seeds.csv', delimiter=',')
```

```python
# Numerization of class type
df_seeds.Clase = df_seeds.Clase.map({'Tipo1' : 0, 'Tipo2': 1, 'Tipo3' : 2})
```

```python
# Normalization
for c in df_seeds.drop(columns=['Clase']).columns:
  min = df_seeds[c].min()
  range = df_seeds[c].max() - min
  df_seeds[c] = (df_seeds[c] - min) / range
```

```python
# Split dataframe into train and test sets
train_set, test_set = train_test_split(df_seeds, test_size=0.2)
```

```python
# examples
X = np.array(train_set.iloc[:, 0:-1])
# colum to predict
Y = np.array(train_set.iloc[:, 7:])
```

```python
# Create the network architecture
model = Sequential()
model.add(Dense(4, input_shape=[7], activation='tanh'))
model.add(Dense(3, activation='sigmoid'))
model.compile(optimizer=SGD(lr=0.2), loss='mse', metrics='accuracy')
```

```python
# Train the neural network
model.fit(X, keras.utils.to_categorical(Y), epochs=400, batch_size=20, verbose=0)
```

```python
print("%% train accuracy : %.2f" % metrics.accuracy_score(
    Y,
    np.argmax(model.predict(X), axis=1))
)
```

```python
print("%% test accuracy : %.2f" % metrics.accuracy_score(
    np.array(test_set.iloc[:, 7:]),
    np.argmax(model.predict(np.array(test_set.iloc[:, 0:-1])), axis=1))
)
```
