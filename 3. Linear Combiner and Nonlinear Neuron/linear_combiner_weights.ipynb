**Ejercicio 6**

El archivo CCPP.csv contiene 9568 datos de una central de ciclo combinado recolectados entre 2006 y 2011. Usando la totalidad de los ejemplos se entrenaron 5 combinadores lineales con el objetivo de predecir el valor del atributo PE correspondiente a la cantidad de energía producida a partir del atributo AT (más detalles de este conjunto de datos aquí). Indique cuáles son los dos combinadores lineales con menor error cuadrático promedio a la hora de predecir el valor de PE en función de AT:

image.png

**a)** W = -2.00 ; b= 500

**b)** W = -2.21 ; b= 498

**c)** W = -2.30 ; b= 497

**d)** W = -2.22 ; b=496.5

**e)** W = -2.16 ; b=496.91


```python
weights = [(-2.00,500), (-2.21, 498), (-2.30, 497), (-2.22, 496.5), (-2.16, 496.91)]
```

```python
#IMPLEMENTO UNA PREDICCION EN BASE A LOS PESOS DADOS
elem_count = len(df_ccpp['AT'])         #cantidad de elementos de entrada
array_of_AT = np.array(df_ccpp['AT'])   #arreglo de elementos AT - temperatura
array_of_PE = np.array(df_ccpp['PE'])   #arreglo de elementos PE - energia
min = 999999999
min_weights = (None, None)
for w in weights:
  sum_ec = 0.0  #esta variable guarda la sumatoria de errores cuadraticos
  w1 = w[0]     #peso de la entrada X_1 o AT
  b = w[1]      #peso de la entrada X_0 (bias)
  for e in range(elem_count):
    y = array_of_AT[e] * w1 + b    #calculo la neta
    ec = (array_of_PE[e] - y)**2   #calculo el error cuadratico
    sum_ec = sum_ec + ec           #sumaria de errores cuadraticos
  mse = sum_ec/elem_count
  if mse < min:
    min = mse
    min_weights = w
print('El combinador lineal con más precisión tenía los pesos:', min_weights)
```
