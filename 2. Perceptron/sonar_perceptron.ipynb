```python
%matplotlib inline
import numpy as np
import pandas as pandas
import time
from sklearn.model_selection import train_test_split
from matplotlib import pylab as plt
from IPython import display
```

```python
#LOAD DATASET
df_sonar = pandas.read_csv('data/sonar.csv')
```

```python
df_sonar['class'].unique()
```

```python
#RANDOM ORDER
df_sonar.sort_values(by=['Atrib_52'], ascending=False)
```

```python
#NUMERIZATION
df_sonar['Mine'] = df_sonar['class'].map({
    'Mine' : 1,
    'Rock' : 0
  })
df_sonar = df_sonar.drop(columns='class')
```

```python
#NORMALIZATION - MEAN & STD
for c in df_sonar.columns:
  if (c != 'Mine'):
    df_sonar[c] = (df_sonar[c] - df_sonar[c].mean()) / df_sonar[c].std()
```

```python
test_set_size = 0.1
for i in range(3):
  #SPLIT INTO TRAIN AND TEST SETS
  df_train, df_test = train_test_split(df_sonar, test_size=test_set_size)
  x1 = df_train.drop(columns='Mine').to_numpy()
  x2 = df_test.drop(columns='Mine').to_numpy()
  t1 = df_train['Mine'].to_numpy()
  t2 = df_test['Mine'].to_numpy()
  accuray_list = []
  for i in range(10):
    perceptron = Perceptron(alpha=0.06, n_iter=200, draw=0, random_state=None)
    #TRAIN PERCEPTRON
    perceptron.fit(x1, t1)
    #TEST PERCEPTRON
    y = perceptron.predict(x2)
    accuracy = 0.0
    accuracy_count = sum(y == t2)
    accuracy = (accuracy_count / len(df_test)) * 100
    accuray_list.append(accuracy)
  print('{0:.2f}% de efectividad,'.format(sum(accuray_list)/10), 'alpha:', perceptron.alpha, '% test set size:', test_set_size * 100)
  test_set_size += 0.1
```
